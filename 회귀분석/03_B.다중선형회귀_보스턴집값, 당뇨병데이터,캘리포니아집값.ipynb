# 보스톤 집값 데이터셋을 이용한 다중선형회귀 모델 구축 및 평가

```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
```

```python
#데이터셋을 로드
boston=load_boston()
```

```python
# 독립변수(피처)와 종속변수를 DataFrame자료형으로 변환
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
# 데이터셋 객체의 data 배열, 즉, 독립변수 X가되는 피처(특성)들을 DataFrame자료형으로 변환하여
# boston_df생성
# boston_df.head() 데이터 5개 확인
```

```python
# 데이터셋 객체의 target배열(boston.target)을 종속 변수인 주택 가격("PRCIE") 컬럼으로 추가
```

```python
boston_df['Price']=boston.target
boston_df.head()
```

# 보스턴 집값 다중선형회귀 분석(사이킷런)

```python
# 데이터 준비, 탐색은 위에서 수행한 것으로 대체
```

### 데이터 준비

#### 데이터 분할

```python
from sklearn.model_selection import train_test_split
X=boston_df.drop('Price',axis=1)
y=boston_df['Price']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#X_train 데이터 일부 확인
X_train.head()
```

### 모델 클래스 선택

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

### 모델 구축 

```python
#학습 수행

reg.fit(X_train,y_train)
```

### 모델 평가

```python
from sklearn.metrics import mean_squared_error, r2_score
# 평가(검증) 데이터로 예측 수행 -> 예측 결과 y_pred구하기
y_pred=reg.predict(X_test)

# 평가 지표 계산
mse = mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)

r2=r2_score(y_test,y_pred)

print("MSE: {0:.3f}, RMSE: {1:.3f}".format(mse,rmse))
print("R2: {:.3f}".format(r2))
```

```python
# 수정된 결정 계수 확인(Ajusted-R2)

#데이터의 크기
n=len(X_train)
#독립변수의 갯수
k=len(X_train.columns)
# 수정된 결정 계수
adj_r2=1-((1-r2)*(n-1)/(n-k-1))
print("Adjusted R2: {:.4f}".format(adj_r2))
```

```python
np.set_printoptions(suppress=True)
print('Y절편: {:.3f}'.format(reg.intercept_))
print('회귀계수: ',np.round(reg.coef_,3))

np.set_printoptions(suppress=False)
```

#### 회귀 계수 확인 : 각각의 독립 변수의 설명력 확인

```python
#회귀  모델에서 구한 회귀 계수와 피처 이름을 묶어 
#시리즈 자료형으로 만듦
coef=pd.Series(data=np.round(reg.coef_,3), index=X.columns)

print('y='+str(reg.intercept_)+' + ')

for i,c in zip(coef.index, coef.values):
    print('('+ str(c)+')*'+str(i))
```

###  실제값과 예측값의 분포차이를 시각화

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))
plt.scatter(y_test,y_pred,s=80)
plt.plot([5,50],[5,50],c='r',ls='--')
plt.xlabel("Actual price ($1,000s)")
plt.ylabel("predicted price")
plt.grid()
plt.show()
```

# 보스턴 집값: 다중선형회귀 분석(스탯츠모델)

```python
# 최소 제곱법으로 객체 생성, 학습, 예측, 평가를 진행한다.
import statsmodels.api as sm

# 훈련 데이터로 학습 수행
X_train=sm.add_constant(X_train)
reg=sm.OLS(y_train,X_train).fit()  #학습 완료
# 검증 데이터로 예측 수행
X_test=sm.add_constant(X_test)
y_pred=reg.predict(X_test)
#평가 지표 값 확인
print(reg.summary())
```

# [응용1]: 다중 선형회귀 연습- load_diabetes

### 필요한 모듈 import

### 데이터 수집

```python
from sklearn.datasets import load_diabetes

diab = load_diabetes()
```

```python
### 데이터 이해
```

```python
dir(diab)
```

### 데이터 준비

```python
# X(독립변수), Y(종속변수) 분할

# 학습용 데이터와 평가용 데이터로 분할하기
```

```python
from sklearn.model_selection import train_test_split
X=diab.data
y=diab.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
### 모델 구축
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

# 회귀 모델 객체 생성
reg=LinearRegression()

#학습 수행
reg.fit(X_train,y_train)

#회귀계수 및 회귀식 완성
coef = pd.Series(data=np.round(reg.coef_,3), index= diab.feature_names)
print(coef)

# 평가(검증) 데이터로 예측 수행 -> 예측 결과 y_pred구하기
y_pred=reg.predict(X_test)


### 모델 평가
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)

r2=r2_score(y_test,y_pred)

n=len(X_train) #샘플의 수
k=len(diab.feature_names)

adj_r2=1-((1-r2)*(n-1)/(n-k-1))

print("MSE:",mse)
print('Y절편:',reg.intercept_)
print('adj_r2:',adj_r2)
```

```python
import statsmodels.api as sm

# 훈련 데이터로 학습 수행
X_train=sm.add_constant(X_train)
reg=sm.OLS(y_train,X_train).fit()  #학습 완료
# 검증 데이터로 예측 수행
X_test=sm.add_constant(X_test)
y_pred=reg.predict(X_test)
#평가 지표 값 확인
print(reg.summary())
```

```python
plt.figure(figsize=(5,5))
plt.scatter(y_test,y_pred,marker='^',s=5)
plt.xlabel("Actual label")
plt.ylabel("Predicted value")
plt.show()
```
