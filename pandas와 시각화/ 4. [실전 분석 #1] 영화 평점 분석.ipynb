# 영화 평점 분석 실습

```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

## 1. 영화 평점 데이터 적재 및 전처리

```python
# 사용자 데이터 읽어오기
users = pd.read_csv('data/movielens/users.dat', sep = '::', engine = 'python',
                   names = ['사용자아이디', '성별','연령','직업','지역'])
users.head()
```

```python
# 평점 데이터 읽어오기
ratings = pd.read_csv('data/movielens/ratings.dat', sep = '::', engine = 'python',
                   names = ['사용자아이디', '영화아이디','평점','타임스탬프'])
ratings.head()
```

```python
# 영화데이터 읽어오기
movies = pd.read_csv('data/movielens/movies.dat', sep = '::', engine = 'python',
                   names = ['영화아이디','영화제목','장르'], encoding = 'latin-1')
movies.head()
```

```python
ratings.사용자아이디.nunique() #평점 한번도 안준 사용자-> 없음
```

```python
#3개의 데이터프레임을 하나로 합치기
data=pd.merge(users,ratings)
data=pd.merge(data, movies)
data.head()
```

## 2. 보고 싶은 영화 찾기
영화들의 평점 평균을 구하여, 사람들에게 인정받는 (평점이 높은) 영화 찾기

```python
# 영화들의 평점 평균을 구하여, 평점이 높은 영화 찾기 
```

```python
영화평점 =data.pivot_table(index='영화제목',values='평점',aggfunc=['mean','count'])
```

```python
영화평점.sort_values(by=('mean','평점'),ascending=False)
```

평균 평점이 만점인 영화들이 최상위에 위치함. 
일반적으로 평점이 만점인 경우는 대부분 평점의 개수가 매우 적은 경우이므로, 이를 확인하기 위해 평점의 개수도 함께 구해본다. 

```python
#평균의 평점 4.3 이상, 평점 개수도 2000개 이상인 영화
영화평점.columns=['평점평균','평점개수']
영화평점[(영화평점.평점평균 >= 4.3) & (영화평점.평점개수 >=2000)]
```

## [실습 #1] 여자들이 좋아하는 영화 찾기 
### - 여성 평점이 4.0 이상이고 여성 평점의 개수가 500개 이상인 영화

```python
data.head()
```

```python
data_f=data[data.성별=='F']
```

```python
ex_f=data_f.pivot_table(index='영화제목',values=['평점'],
                 aggfunc=['mean','count'])
ex_f.columns=['평점','개수']
ex_f.head()
```

```python
여성인기영화 = ex_f[(ex_f.평점 >=4.0) & (ex_f.개수>=500)]
여성인기영화
```

## [실습 #2] 실습 #1에서 구한 영화(여성인기영화)의 장르를 분석해 보자.
여성인기영화의 장르 통계 구하기

예를 들어, 여성인기영화 중 Drama 장르의 영화는 10개, Action 영화는 3개, ...

```python
여성인기영화.index
```

```python
#isin() 함수 사용해서 movies 에서 여성인기영화의 장르만 선택
여성인기영화장르 = movies[movies.영화제목.isin(여성인기영화.index)].장르
여성인기영화장르
```

```python
여성인기영화장르 = 여성인기영화장르.str.split('|',expand=True)
여성인기영화장르
```

```python
여성인기영화장르[0].value_counts().add(여성인기영화장르[1].value_counts() , fill_value=0)
```

```python
여성인기장르 = Series(dtype='float64')
for col in 여성인기영화장르.columns:
    여성인기장르 =여성인기장르.add(여성인기영화장르[col].value_counts(),fill_value=0)
```

```python
여성인기장르.sort_values(ascending=False)
```

## [실습 #3] 남자와 여자의 호불호가 크게 갈리는 영화 10개 찾기
전체 평점의 개수가 500개 이상인 영화만 대상으로 함.

```python
ex=data.pivot_table(index='영화제목',values='평점',
                 aggfunc=['mean','count'])
ex.columns=['평점','개수']
ex[ex.개수>=500]
```

```python
a=(ex[ex.개수>=500].index).tolist()
movie=data[data['영화제목'].isin(a)]
```

```python
movie.head()
```

```python
fem_movie=movie[movie.성별=='F'].pivot_table(index='영화제목',values=['평점'],
                 aggfunc=['mean','count'])
fem_movie.columns=['평점_f','개수']
fem_movie.head()
```

```python
mal_movie=movie[movie.성별=='M'].pivot_table(index='영화제목',values=['평점'],
                 aggfunc=['mean','count'])
mal_movie.columns=['평점_m','개수']
mal_movie.head()
```

```python
diff=(abs(fem_movie.평점_f - mal_movie.평점_m)).sort_values(ascending=False)
```

## [실습 #4] 연령대 별로 영화 평점 분석하기
연령대(10대 미만, 10대, 20대, ...50대) 컬럼을 추가한 후, 영화별 연령대별 영화평점 구하기

```python
data.head()
```

```python
#연령의 시작하는 자리수에 따라 연령대 생성
#한자리수이면 10대 미만 , 아닌경우는 시작자리수에 따라 결정
#data=data.astype({'연령': str})
#data['연령대']='1'
data.연령 >=10
def age_category(age):
    age=str(age//10)+'대'
    return age
연령대=data.연령.apply(age_category)
data['연령대']=연령대
#data['연령대']=a
#data['연령대']=pd.DataFrame(data['연령'])
data
```
