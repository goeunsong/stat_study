##### 순차 패턴탐사
서로 다른 항목들 간의 발생순서를 발견하는 기법

장바구니 분석에서 어떤 고객의 시간에 따른 구매 정보를 활용함으로써 구매 순서가 고려된 연관성 발견

##### 시퀀스 데이터베이스
이벤트들이 순서대로 들어감

##### 순차 패턴
순차성 탐사를 통해 발견되는 패턴들

이벤트들 X가 발생한 이후에 이벤트 Y 가 발생한다는 순차성

##### 순차 패텀 탐사 과정
1) sort phase : 행위자 ID와 시간 순서에 따라 원본 데이터 변환/정렬

2) large itemset and transform phase : 빈발 항목집합을 탐사해 시퀀스 형태로 변환

3) sequence phase : 빈발 항목집합을 이용해 빈발한 시퀀스들 찾기

4) maximal phase : (필요한 경우) 빈발한 시퀀스 집합 중 가장 길이가 긴 시퀀스 추출

```python
import prefixspan
#help(prefixspan)
```

```python
seq=[['맥주','오징어','맥주','땅콩','소주'],
    ['과자','오징어','우유','땅콩','커피','계란'],
    ['맥주','오징어','땅콩','라면','맥주','휴지','땅콩','과자'],
    ['계란','우유','오징어','휴지','라면']]
```

```python
import prefixspan as ps
result=ps.PrefixSpan(seq)
```

```python
#가장 빈발한 상위 k개의 순차 패턴 찾기
result.topk(5)
```

```python
#filter 에는 특정 조건을 만족하는 결과만 확인하도록 lambda 함수 적용
#patt: 빈도 matches: 패턴
result.topk(5,filter=lambda patt, matches: len(patt)>1)
```

```python
result.topk(5,filter=lambda patt, matches: len(matches)>2)
```

```python
#지정한 최소 지지도(횟수) 이상으로 발생한 순차 패턴 찾기
result.frequent(2)
```

```python
result.frequent(2,filter=lambda patt,matches: len(patt)>2)
```

```python
import csv
seq=[]
with open('market.csv') as f:
    rd=csv.reader(f)
    for r in rd:
        seq.append(r)
```

```python
for s in seq:
    print(s)
```

```python
result=ps.PrefixSpan(seq)
result.topk(5,filter=lambda patt, matches: len(patt)>1)
```

```python
result.frequent(200,filter=lambda patt, matches:len(patt)>2)
```
