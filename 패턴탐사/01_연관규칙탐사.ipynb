##### 연관 규칙 탐사
연관 규칙 : 연관성 탐사를 통해 발견되는 규칙

항목들 X가 발생하였을 때 항목들 Y도 같이 발생. ( X -> Y )

1) X 와 Y 는 상호배타적인 항목들의 집합

2) X를 선행절/조건절 , Y를 후행절/결과절

##### 트랜잭션
데이터베이스 관리 측면에서는 완결성이 있는 1회의 논리적 실행 또는 처리 단위

연관 규칙 탐사 에서는 1개의 거래이력 단위

트랜잭션 내에 들어있는 각각의 개별적인 자료를 항목이라 함 + 항목들 간의 순서는 고려되지 않음

##### 항목집합
트랜잭션 내에 들어있는 항목들의 집합

#### 지지도
전체 트랜잭션들에서 특정 항목집합이 포함되어 있는 확륙 (발생 빈도 비율)

#### 빈발 항목집합
전체 트랜잭션들에서 지지도가 특정 임계값(=최소 지지도) 이상으로 빈발하게 발생한 항목집합

#### 신뢰도
연관 규칙의 선행절이 포함된 트랜잭션들 중 후행절의 항목 집합들이 포함될 확률

#### 결과
빈발 항목집합이면서 신뢰도가 특정 임계값(=최소 신뢰도) 이상인 연관 규칙들을 의미있는 결과로 간주

#### 향상도
연관 규칙의 선행절과 후행절이 독립인 경우 대비, 선행절과 후행절이 함께 발생할 확률 (신뢰도 / 지지도)

```python
#트랜잭션의 총 수
tr_num=50

#한 트랜잭션 내의 항목 개수(범위)
tr_len=list(range(2,6))

#항목이 가질 수 있는 값의 범위
item_domain = list(range(1,21))
```

```python
print(item_domain)
```

```python
data = []

for a in range(tr_num):
    c=random.sample(item_domain,random.choice(tr_len))
    c.sort()
    data.append(c)
```

```python
data
```

```python
import apyori as ap
rules= list(ap.apriori(data, min_support=0.1,min_confidence=0.2))
rules
```

```python
for r in rules:
    if len(r[0])>1:
        support = r[1]
        ar = list(r[2])
        for a in ar:
            print(a[0],'==>',a[1])
            print('support:',r[1])
            print('confidence:',a[2])
            print('list:',a[3])
            print()
```

```python
import pandas as pd

result = []

for r in rules:
    if len(r[0])>1:
        support=r[1]
        ar=list(r[2])
        for a in ar:
            row=[set(a[0]),'==>',set(a[1]),support,a[2],a[3]]
            result.append(row)
```

```python
result
```

```python
col=['선행절','==>','후행절','지지도','신뢰도','향상도']
df=pd.DataFrame(result,columns=col,index=range(1,len(result)+1))
df
```

```python
#연관규칙탐사
import apyori as ap
f= open('market.csv')
tr = list(ap.load_transactions(f,delimiter=','))
f.close()
```

```python
for t in tr:
    print(t)
```

```python
result = ap.apriori(tr,min_support=0.05,min_confidence=0.1)
rules=list(result)
for r in rules:
    if len(r[0])>1:
        support = r[1]
        for i in list(r[2]):
            print(set(i[0]),'==>',set(i[1]))
            print('지지도:',round(r[1],4))
            print('신뢰도:',round(i[2],4))
            print('향상도:',round(i[3],4))
            print()
            
```

```python
#min_support 를 0.01 로 변경
result = ap.apriori(tr,min_support=0.01,min_confidence=0.1)
rules=list(result)
for r in rules:
    if len(r[0])>1:
        support = r[1]
        for i in list(r[2]):
            print(set(i[0]),'==>',set(i[1]))
            print('지지도:',round(r[1],4))
            print('신뢰도:',round(i[2],4))
            print('향상도:',round(i[3],4))
            print()
            
```

```python
import pandas as pd

result = []

for r in rules:
    if len(r[0])>1:
        support=r[1]
        ar=list(r[2])
        for i in ar:
            row=[set(i[0]),'==>',set(i[1]),support,i[2],i[3]]
            result.append(row)
```

```python
col=['선행절','==>','후행절','지지도','신뢰도','향상도']
df=pd.DataFrame(result,columns=col,index=range(1,len(result)+1))
df
```
